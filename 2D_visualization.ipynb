# Exploring Data - Visualization

```python
import sklearn
from sklearn.cluster import KMeans 
import matplotlib.pyplot as plt
```

```python
cluster_amount = 5
```

```python
mstd = [] 
m_total_transactions = [] 
mtransf = [] 
mean_ISP = [] 
mean_border = [] 

transf_std = [] 
transf_total_transactions = [] 
transf_ISP = [] 
transf_border = [] 

ISP_std = [] 
ISP_total_transactions = [] 
ISP_border = [] 

total_transactions_std = [] 
total_transactions_border = [] 

border_std = [] 

for name in sender.iteritems(): 
    account_num = name[0]
    value = name[1]
    
    total_transaction = value['transactions']
    mean_trans = value['mean_transaction_amount']
    std_trans = value['transacton_std_dev']
    trans_frequency = value['transaction_frequency']
    senderISP = value['total_ISPs']
    distance_us = value['distance_from_border']
    
    mstd.append([mean_trans,std_trans])
    m_total_transactions.append([mean_trans,total_transaction])
    mtransf.append([trans_frequency,mean_trans])
    mean_ISP.append([senderISP,mean_trans])
    mean_border.append([distance_us,mean_trans])
    
    transf_std.append([trans_frequency,std_trans])
    transf_total_transactions.append([trans_frequency,total_transaction])
    transf_ISP.append([senderISP,trans_frequency])
    transf_border.append([distance_us,trans_frequency])
    
    ISP_total_transactions.append([senderISP,total_transaction])
    ISP_std.append([senderISP,std_trans])
    ISP_border.append([senderISP,distance_us])
    
    total_transactions_std.append([total_transaction,std_trans])
    total_transactions_border.append([total_transaction,distance_us]) 

    border_std.append([distance_us,std_trans])

    
all_features = [mstd, m_total_transactions, mtransf, mean_ISP, mean_border, transf_std, transf_total_transactions,
               transf_ISP, transf_border, ISP_std, ISP_total_transactions,ISP_border,total_transactions_std, 
               total_transactions_std, total_transactions_border, border_std] 

for i in range(len(all_features)): 
    all_features[i] = sklearn.preprocessing.scale(all_features[i])
```

```python
#print(np.shape(cluster_data_mstd))
kmeans = KMeans(n_clusters=cluster_amount, random_state=0).fit(all_features[0])
centers = kmeans.cluster_centers_
labels = kmeans.labels_
np.shape(centers)

x = [] 
y = []

center_x = [] 
center_y = [] 

for point in all_features[0]: 
    x.append(point[0])
    y.append(point[1])
    
for point in centers: 
    center_x.append(point[0])
    center_y.append(point[1])
    
plt.scatter(x,y,c=labels.astype(float))
plt.scatter(center_x,center_y,marker="x",s=200)
plt.title('Mean Transaction Amount vs. Spontenaity in Transaction Amount')
plt.ylabel('Transaction Amount Standard Deviation')
plt.xlabel('Mean Transaction Amount')
plt.show()
```

# Final Figures for Presentation - Sender

```python
data0 = []
data1 = [] 
data2 = [] 
for name in sender.iteritems(): 
    account_num = name[0]
    value = name[1]
    
    total_transaction = value['transactions']
    std_trans = value['transacton_std_dev']
    mean_trans = value['mean_transaction_amount']
    freq = value['transaction_frequency']
    senderISP = value['total_ISPs']
    city = value['major_city']
    night = value['night_time']
    
    data0.append([senderISP,night])
    data1.append([total_transaction,std_trans])
    data2.append([city,mean_trans,freq])
    
final_features = [data0,data1,data2]
for i in range(len(final_features)): 
    final_features[i] = sklearn.preprocessing.scale(final_features[i])
```

```python
#print(np.shape(mean_border))
kmeans = KMeans(n_clusters=cluster_amount, random_state=0).fit(final_features[2])
centers = kmeans.cluster_centers_
labels = kmeans.labels_
np.shape(centers)

x = [] 
y = []
z = [] 

center_x = [] 
center_y = [] 
center_z = [] 

for point in final_features[2]: 
    x.append(point[0])
    y.append(point[1])
    z.append(point[2])
    
for point in centers: 
    center_x.append(point[0])
    center_y.append(point[1])
    center_z.append(point[2])

fig=plt.figure(figsize=(8, 6), dpi= 80, facecolor='w', edgecolor='k')
ax = plt.axes(projection='3d')
ax.scatter3D(x, y, z, c=labels.astype(float))
ax.scatter3D(center_x,center_y,center_z,marker="x",s=200)
plt.title('Clustering in 3D - Distance from Major City vs. Mean Transaction Amount vs. Frequency of Transactions')
plt.xlabel('Distance from Major City')
plt.ylabel('Mean Transaction Amount')
ax.set_zlabel('Transaction Frequency')
plt.show()
```

```python
accounts = sender.T
# TODO: get subset of 10,000 accounts
accounts.rename(columns={'transacton_std_dev': 'transaction_std_dev'}, inplace=True)
accounts.head()

data10D = accounts[["transactions", "mean_transaction_amount", "transaction_std_dev", "mean_trans_times", "trans_times_std", "total_ISPs", "distance_from_border", "unique_receivers", "night_time", "major_city"]]
data10D_scaled = sklearn.preprocessing.scale(data10D)

# run clustering on the 10D data, but plot only 2 features
num_clusters = 15

# feature indices
i = 5
j = 8



kmeans = KMeans(n_clusters=num_clusters, random_state=0).fit(data10D_scaled)
centers = kmeans.cluster_centers_
labels = kmeans.labels_
print(np.shape(centers))


center_x = []
center_y = []

   
for point in centers:
    center_x.append(point[i])
    center_y.append(point[j])

fig = plt.figure(figsize=(8,6),dpi=80, facecolor='w', edgecolor='k')
   
plt.scatter(data10D_scaled[:,i], data10D_scaled[:,j], c=labels.astype(float), s=10)

plt.title("Clustering on 10-D Data - Number if ISPs vs. Number of transactions sent from 10pm - 6am")
plt.xlabel('Number of ISPs')
plt.ylabel('Number of Transactions sent from 10pm - 6am')
plt.show()

fig = plt.figure(figsize=(8,6),dpi=80, facecolor='w', edgecolor='k')


plt.scatter(center_x,center_y, marker="x",s=100, color="black", alpha=0.8)

plt.title("Clustering on 10-D Data - Number if ISPs vs. Number of transactions sent between 10pm - 6am")
plt.show()
```
